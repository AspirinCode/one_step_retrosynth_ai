```python
import pandas as pd

from data_methods.dataset_construction import extract_relevant_information
from chem_methods.reaction_cores import get_reaction_core_atoms
from chem_methods.drawing import draw_molecule, draw_reaction, draw_fingerprint_substructures
```

```python
source_dataset = pd.read_csv("D:/Datasets/uspto_liu_coley2/data_processed.csv")[["id", "rxn_smiles", "class"]]
reactant_pool = pd.read_pickle("D:/Datasets/one_step_retrosynth_ai_output/unique_reactants_pool.pkl")["reactant_canonical_smiles"].values.tolist()
product_pool = pd.read_pickle("D:/Datasets/one_step_retrosynth_ai_output/unique_products_pool.pkl")["product_canonical_smiles"].values.tolist()
fp_params = {"type": "ecfp", "radius": 1, "bits": 1024}
ctr = 0

for ind, row in source_dataset.iterrows():
    
    if row["class"] != 5:
        continue
        
    print("Generating unique molecules mapping for row {}.".format(ind))
    ctr += 1
    
    reaction_cores = get_reaction_core_atoms(row["rxn_smiles"])
    display(draw_reaction(row["rxn_smiles"], reaction_cores=reaction_cores))
    
    lol = extract_relevant_information(row["rxn_smiles"], reactant_pool, product_pool, fp_params)
    
    for l in lol[3]:
        display(draw_molecule(l))
    
    print("***********************************")
    
    for l in lol[7]:
        display(draw_molecule(l))
        
    print("***********************************")
        
    #for l in lol[12]:
        #display(draw_molecule(l))
    
    #print("***********************************")
        
    #for l in lol[16]:
        #display(draw_molecule(l))
    
    print("---------------------------------------------------------------------------------------------------------------")
    
    if ctr > 10: 
        break
```
